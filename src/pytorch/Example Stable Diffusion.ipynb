```python
model_id = "runwayml/stable-diffusion-v1-5"

from diffusers import StableDiffusionPipeline                                                                                                                                                                                                 
pipe = StableDiffusionPipeline.from_pretrained(model_id) 

import torch      

#prompt = "portrait photo of a old warrior chief" #  in green
# prompt = "photorealistic micky mouse tron star wars lego" 

torch_device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(torch_device)
```

```python
generator = torch.Generator(torch_device).manual_seed(7)
prompt = "A photograph of a white mexican mouse." 
image = pipe(prompt, generator=generator).images[0]
image
```

```python
for x in range(10):
    generator = torch.Generator(torch_device).manual_seed(x)
    prompt = "photorealistic battlestar lego" 
    image = pipe(prompt, generator=generator).images[0]
    image.save('./images/bs_image_{}.png'.format(x))
```
